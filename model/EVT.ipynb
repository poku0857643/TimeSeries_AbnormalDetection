```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genpareto
```

# Prepare data (e.g. daily stock returns)

```python
data = np.random.exponential(scale=2, size=1000)
```

Set a threshold (e.g. the 99th percentile)

```python
threshold = np.percentile(data, 99)
exceedances = data[data > threshold] - threshold
```

# Fit the GPD

```python
shape, loc, scale = genpareto.fit(exceedances)
```

Calculate a 99.9% Value at Risk (VaR)

```python
n = len(data)
n_u = len(exceedances)
p = 0.0001 # 99.9% confidence
var_evt = threshold + (scale / shape) * (((n/n_u) * p)** (-shape) - 1)
print(f"Shape Paremeter: {shape:.2f}")
print(f"99.9% VaR: {var_evt:.2f}")
```

# Visualization

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, alpha=0.7, color="skyblue", label="Data Distribution")
plt.axvline(var_evt, color="red", linestyle="--", linewidth=2, label=f"99.9% VaR: {var_evt:.2f}")
plt.title("Data Distribution and Extreme Value Theory VaR")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
print("Data:", data[:10], "...")
print(f"VaR: {var_evt:.2f}")
```
